```python
import os
import pandas as pd
from datetime import datetime

keyInverstorFilePath = '/Users/fang/PycharmProjects/PressPlay/keyInvestorMapping.csv'
keyInvestorDf = pd.read_csv(keyInverstorFilePath)

dateStr = datetime.now().strftime('%Y%m%d')
stockDataFilePath = '/userap/BuySellReport/bs_data/' + dateStr

for row in keyInvestorDf.iterrows():
    stockNo = row[1]['股票代號']
    keyBrokerNo = row[1]['券商代號']
    filePath = None
    if os.path.exists(stockDataFilePath + os.sep + 'twse' + os.sep + stockNo + '.csv'):
        filePath = stockDataFilePath + os.sep + 'twse' + os.sep + stockNo + '.csv'
    elif os.path.exists(stockDataFilePath + os.sep + 'tpex' + os.sep + stockNo + '.csv'):
        filePath = stockDataFilePath + os.sep + 'tpex' + os.sep + stockNo + '.csv'
    if not filePath:
        print("file doesn't exist")
    else:
        stockDf = pd.read_csv(filePath)
        if keyBrokerNo in stockDf['券商'].to_list():
            print(row[1]['股票名稱'], ':', row[1]['券商名稱'])
```
