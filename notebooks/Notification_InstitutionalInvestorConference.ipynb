```python
import os
import logging
import pandas as pd
from datetime import datetime, timedelta
```

```python
root_path = './'
start_dt = datetime.strptime(datetime.now().strftime("%Y-%m-%d"), "%Y-%m-%d")
end_dt = start_dt + timedelta(days=8)
start_year = start_dt.strftime("%Y")
end_year = end_dt.strftime("%Y")
logging.info(f'start: {start_dt}')
logging.info(f'end_dt: {end_dt}')

pdf_list = []
for year in set([start_year, end_year]):
    file_name = f'institutionalInvestorConference_{year}.csv'
    pdf = pd.read_csv(root_path + os.sep + file_name)
    pdf['法說會日期'] = pd.to_datetime(pdf['法說會日期'])
    sub_pdf = pdf[(pdf['法說會日期']>start_dt) & (pdf['法說會日期']<end_dt)]
    pdf_list.append(sub_pdf)
result_pdf = pd.concat(pdf_list, axis=0, ignore_index=True, sort=False)
result_pdf = result_pdf.sort_values(['法說會日期', '股票代號'])[['股票代號', '法說會日期', '地點', '擇要訊息']]
result_pdf
```
